1. iterations means loops to work with repeative tasks or repeatative performances
2. python have 2types of iterators
3. `for` loop
4. `while` loop


````python
#write a program to iterate name using loop
name="shivaji"
for char in name:
    print(char)
````

````python
# write a program to find the even numbers in between 100 to 150 using loop
for n in range(100,151):
    if(n%2==0):
        print(n,end=",")
````

````python
# write a program to find the even numbers in between 100 to 150 using loop
for n in range(100,151):
    if(n%1==0):
        print(n,end=",")
````

````python
# wrie a program to find the vowel from string using for loop
name=str(input("please enter your name here:"))
for char in name:
    if (char in "aeiouAEIOU"):
        print(char)
````

````python
# wrie a program to find the non vowel from string using for loop
name=str(input("please enter your name here:"))
for char in name:
    if (char not in "aeiouAEIOU"):
        print(char)
````

````python
# write a program to read name from list using for loop
names=["vikas","shivaji",'aditya','chandra']
for n in names:
    print("my name is",n )
````
